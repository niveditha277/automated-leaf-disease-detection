```python
import tkinter as tk
from tkinter import filedialog, Label, Button, Frame
from PIL import Image, ImageTk
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import os

model_path = r"C:\Users\Vanshika\OneDrive\Desktop\niveditha files\automated leaf disease detection\model.h5"


if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model file not found at: {model_path}")


model = load_model(model_path)

class_labels = ['Healthy', 'Powdery Mildew', 'Downy Mildew', 'Bacterial Blight', 'Rust']

def predict_disease(image_path):
    try:
        img = image.load_img(image_path, target_size=(28, 28)) 
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0  
        prediction = model.predict(img_array)
        predicted_class = np.argmax(prediction, axis=1)[0]
        confidence = np.max(prediction) * 100
        return class_labels[predicted_class], confidence
    except Exception as e:
        return f"Error: {str(e)}", 0

def open_image():
    global img_label, img_tk  
    file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png")])
    if file_path:
        img = Image.open(file_path)
        img = img.resize((200, 200))
        img_tk = ImageTk.PhotoImage(img) 
        img_label.config(image=img_tk)
        img_label.image = img_tk  

        result, confidence = predict_disease(file_path)
        result_label.config(text=f"Disease: {result}")
        confidence_label.config(text=f"Confidence: {confidence:.2f}%")

root = tk.Tk()
root.title("Leaf Disease Detection")
root.geometry("500x600")
root.configure(bg="#f0f0f0")

title_label = Label(root, text="Leaf Disease Detection", font=("Arial", 24, "bold"), bg="#f0f0f0")
title_label.pack(pady=20)

image_frame = Frame(root, bg="#ffffff", bd=2, relief="groove")
image_frame.pack(pady=10, padx=20, fill="both", expand=True)

img_label = Label(image_frame, text="No Image Selected", width=25, height=10, bg="lightgray")
img_label.pack(pady=10)

open_button = Button(root, text="Upload Leaf Image", command=open_image, font=("Arial", 16), bg="#4CAF50", fg="white")
open_button.pack(pady=20)

result_label = Label(root, text="Disease: ", font=("Arial", 16), bg="#f0f0f0")
result_label.pack(pady=10)

confidence_label = Label(root, text="Confidence: ", font=("Arial", 16), bg="#f0f0f0")
confidence_label.pack(pady=10)
root.mainloop()
```
